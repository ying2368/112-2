1.假設有一筆資料scores = {'Math':[90,np.nan,70,80],
                 'English':[60,70,'90',50],
                 'History':[33,np.nan,188,60]}
請先將它匯入至df df = pd.DataFrame(scores, index=['Simon','Allen','Jimmy','Peter'])
a.請將English裡有問題的’90’改為數字90
b.請將History裡的188，改成88
c.請將遺漏值nan，設為各自欄位的平均值(提示用fillna)
d.請確認資料型態

```python
import pandas as pd
import numpy as np

scores = {'Math': [90, np.nan, 70, 80],
      'English': [60, 70, '90', 50],
      'History': [33, np.nan, 188, 60]}
df = pd.DataFrame(scores, index=['Simon', 'Allen', 'Jimmy', 'Peter'])
df
```

a.請將English裡有問題的’90’改為數字90

```python
df['English'] =df['English'].replace('90', 90)
df['English']
```

b.請將History裡的188，改成88

```python
df['History'] = df['History'].replace(188, 88)
df['History']
```

c.請將遺漏值nan，設為各自欄位的平均值(提示用fillna)

```python
df = df.fillna(df.mean())
df
```
