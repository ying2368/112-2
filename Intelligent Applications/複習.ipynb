##Ch1 AI-python

````python
# 8
def cal_price_dict(k_cost):
    rate = 0.03
    nt_cost = k_cost * rate
    inc = 0.2
    nt_price = nt_cost * (1 + inc)
    data = {
        'k_cost': k_cost,
        'nt_cost': nt_cost,
        'nt_price': nt_price
    }
    return data
cal_price_dict(3000)
````

````python
# 10
list_k_cost = [30000, 500, 45000, 50000]
list_results = []
for k_cost in list_k_cost:
    if k_cost > 10000:
        data = cal_price_dict(k_cost)
        list_results.append(data)
list_results
````

````python
# 11
import pandas as pd
df = pd.DataFrame(list_results)
df.columns = ['韓元成本','台幣成本','台幣售價']
df.mean()
````

````python
df.mean(axis=1)
````

````python
df.max()
````
